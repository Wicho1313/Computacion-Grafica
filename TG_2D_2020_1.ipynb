1. Construya una función que le permita escalar una figura en dos dimensiones. Usted deberá redimensionar la figura original a una más grande o una más pequeña con un valor de escalado aleatorio. La función debe recibir por argumentos el conjunto de puntos que representa su figura y un valor que indique que matriz de transformación de escalado va a utilizar. Grafique la fingura orginal y la escalada.

```
import numpy as np
import matplotlib.pyplot as plt
#Función que permite escalar una figura en dos dimensiones.
#Recibe: x=Conjunto de puntos x de la figura del pato
#        y=Conjunto de puntos y de la figura del pato
#        h=Arreglo de 1's del tamaño de x 
#        e=Selección del usuario
#Devuelve: pp=Matriz de la figura escalada.

def escalar_figura(x,y,h,e):
  #Lista de todos los conjuntos de puntos
  P=np.array([x,y,h])
  if e=='normal':
    pp=np.dot(TI,P)
    pp=np.dot(S,pp)
    pp=np.dot(T,pp)
  elif e=='inversa':
    pp=np.dot(TI,P)
    pp=np.dot(SI,pp)
    pp=np.dot(T,pp)
  else:
    print('Coloque "normal" o "inversa"')

  return pp


e=input('Escribir si quiere "normal" o "inversa": ')
#se genera la figura
x=np.array([4,8,4,4,2,4,2,2,4,6,7,5,2,6,5,5,5,4,2,10,7])
y=np.array([1,1,5,1,3,3,3,5,5,7,7,9,7,7,7,9,7,7,5,5,2])
h=np.ones(len(x))
  #Se genera valor de escalado con numeros aleatorios entre 5 y 10
sx=np.random.randint(5,10)
sy=sx
#matriz de escalado normal
S=np.array([[sx,0,0],
              [0,sy,0],
              [0,0,1]])
#matriz de escalado inversa
SI=np.array([[1/sx,0,0],
             [0,1/sy,0],
             [0,0,1]])
  
#Matríz de traslación
tx=x[0]
ty=y[0]

T=np.array([[1,0,tx],
            [0,1,ty],
            [0,0,1]])
#Matríz de traslación inversa
TI=np.array([[1,0,-tx],
             [0,1,-ty],
             [0,0,1]])
#se llama a la función
PP=escalar_figura(x,y,h,e)
#se grafica la figura y la figura escalada
plt.plot(x,y,color='g')
plt.plot(PP[0,:],PP[1,:],color='m')

#Definición de figura
plt.grid(True)
plt.show()
```

2. Construya una función que le permita trasladar una figura en dos dimensiones. Usted deberá mover la figura original entre los diferentes cuadrantes de un espacio en dos dimensiones. 

 * Grafique los cuadrantes.
 * La posición inicial de la figura deberá ser en el cuadrante $-xy$. 
 * La función debe recibir por argumentos el conjunto de puntos que representan a la figura.
 * Grafique la figura orginal y las diferentes traslaciones entre los cuadrantes.
 * Las figuras generadas deben tener un color diferente.


```
#Función que me permite trasladar una figura en dos dimensiones
#Recibe: El conjunto de puntos 'x', 'y' y una matriz de 1's 
def trasladar(o,tx,ty,col):
  T=np.array([[1,0,tx],
            [0,1,ty],
            [0,0,1]])
  pp2=np.dot(T,o)
  plt.plot(pp2[0,:],pp2[1,:],color=col)
  
plt.plot(x,y,color='y')
#Origen, conjunto de puntos
o=np.array([x,y,h])
#graficando en el cuadrante xy
trasladar(o,0,-10,'r')
trasladar(o,-11,0,'b')
trasladar(o,-11,-10,'m')

#Definición de figura y gráfico
plt.grid(True)
plt.axvline(0,0,color='black')
plt.axhline(0,0,color='black')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.title('Traslación',c='r')
plt.xlabel('Eje X',c='r')
plt.ylabel('Eje Y',c='r')
plt.show()
```

5. Construya una función que le permita hacer reflexión de una figura en dos dimensiones sobre el eje $x$ y sobre el eje $y$. 

 * La posición inicial de la figura deberá ser en el cuadrante xy positivo. 

 * La función debe recibir por argumentos el conjunto de puntos que representa  su figura.
 * La función debe graficar la reflexión de la figura en los diferentes cuadrantes. 

 * Grafique la figura original y los resultados de la reflexión.


```
#Función que hace la reflexión en los diferentes cuadrantes.
#Recibe: o = Conjunto de puntos de la figura
def reflexion(o):
  m=np.dot(rx,o)
  plt.plot(m[0,:],m[1,:],color='m')

  m=np.dot(ry,m)
  plt.plot(m[0,:],m[1,:],color='r')

  m=np.dot(rx,m)
  plt.plot(m[0,:],m[1,:],color='black')

#Matriz de rotación en eje x
rx=np.array([[1,0,0],
             [0,-1,0],
             [0,0,1]])
#Matriz de rotación en eje y
ry=np.array([[-1,0,0],
             [0,-1,0],
             [0,0,1]])
#Posicion inicial en xy positivo
plt.plot(o[0,:],o[1,:],color='g')
#Se hace llamado a función
reflexion(o)
#Definición de figura y gráfico
plt.grid(True)
plt.axvline(0,0,color='black')
plt.axhline(0,0,color='black')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.title('Reflexión',c='r')
plt.xlabel('Eje X',c='r')
plt.ylabel('Eje Y',c='r')
plt.show()
```
