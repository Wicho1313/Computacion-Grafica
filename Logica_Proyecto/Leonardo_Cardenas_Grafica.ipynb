```python
def lons_lat_dataframe(df1, df2, df3, df4, df5, df6, df7, df8):
    
    for i in range (len(df1)):
        #codigo dataframe el nombre de la columna
        nombre_estado = df1['ENTIDAD_UM'].iloc[i]
        sector = df1['SECTOR'].iloc[i]
        sexo = df1['SEXO'].iloc[i]
        paciente = df1['TIPO_PACIENTE'].iloc[i]
        intubado = df1['INTUBADO'].iloc[i]
        neumonia = df1['NEUMONIA'].iloc[i]
        diabetes = df1['DIABETES'].iloc[i]
        epoc = df1['EPOC'].iloc[i]
        asma = df1['ASMA'].iloc[i]
        inmusupr = df1['INMUSUPR'].iloc[i]
        hipertension = df1['HIPERTENSION'].iloc[i]
        otra = df1['OTRA_CON'].iloc[i]
        cardiovascular = df1['CARDIOVASCULAR'].iloc[i]
        obesidad = df1['OBESIDAD'].iloc[i]
        renal = df1['RENAL_CRONICA'].iloc[i]
        tabaquismo = df1['TABAQUISMO'].iloc[i]
        otro = df1['OTRO_CASO'].iloc[i]
        resultado = df1['RESULTADO'].iloc[i]
        cui = df1['UCI'].iloc[i]
        nacionalidad = df1['NACIONALIDAD'].iloc[i]
        
        for j in range (len(df2)):  
            # codigo del dataframe estados
            cod_entidad = int(df2['CLAVE_ENTIDAD'].iloc[j])
            if (nombre_estado == cod_entidad):
                df1['estado_um'].iloc[i] = df2['ENTIDAD_FEDERATIVA'].iloc[j] 
                df1['estado_res'].iloc[i] = df2['ENTIDAD_FEDERATIVA'].iloc[j] 
                df1['longitud'].iloc[i] = df2['LONGITUD'].iloc[j] 
                df1['latitud'].iloc[i] = df2['LATITUD'].iloc[j]                 
                break
                
        for a in range (len(df3)):
            # sector 
            cod_sector = int(df3['CLAVE'].iloc[a])
            if (sector == cod_sector):
                        df1['sector'].iloc[i] = df3['DESCRIPCION'].iloc[a]
                        break                
        for b in range (len(df4)):
            #sexo
            cod_sexo = int(df4['CLAVE'].iloc[b])
            if (sexo == cod_sexo):
                df1['sexo'].iloc[i] = df4['DESCRIPCION'].iloc[b]
                break
                
        for c in range (len(df5)):
            # paciente
            cod_paciente = int(df5['CLAVE'].iloc[c])
            if (paciente == cod_paciente):
                df1['tipo_paciente'].iloc[i] = df5['DESCRIPCION'].iloc[c]
                break
                
        for d in range (len (df6)):
            cod_sino = int(df6['CLAVE'].iloc[d])
            if (intubado == cod_sino):
                df1['intubado'].iloc[i] = df6['DESCRIPCION'].iloc[d]
                break
                
        for e in range (len (df6)):
            cod_sino = int(df6['CLAVE'].iloc[e])
            if (neumonia == cod_sino):
                df1['neumonia'].iloc[i]  = df6['DESCRIPCION'].iloc[e]
                break
                
        for f in range (len (df6)):
            cod_sino = int(df6['CLAVE'].iloc[f])
            if (diabetes == cod_sino):
                df1['diabetes'].iloc[i] = df6['DESCRIPCION'].iloc[f]
                break
                
        for g in range (len (df6)):
            cod_sino = int(df6['CLAVE'].iloc[g])
            if ( epoc== cod_sino):
                df1['epoc'].iloc[i] = df6['DESCRIPCION'].iloc[g]
                break
                
        for h in range (len (df6)):
            cod_sino = int(df6['CLAVE'].iloc[h])
            if ( asma == cod_sino):
                df1['asma'].iloc[i] = df6['DESCRIPCION'].iloc[h]
                break
                
        for k in range (len (df6)):
            cod_sino = int(df6['CLAVE'].iloc[k])
            if ( inmusupr== cod_sino):
                df1['inmusupr'].iloc[i] = df6['DESCRIPCION'].iloc[k]    
                break
                
        for l in range (len (df6)):
            cod_sino = int(df6['CLAVE'].iloc[l])
            if (hipertension == cod_sino):
                df1['hipertension'].iloc[i] = df6['DESCRIPCION'].iloc[l]    
                break
                
        for m in range (len (df6)):
            cod_sino = int(df6['CLAVE'].iloc[m])
            if ( otra== cod_sino):
                df1['otra_con'].iloc[i] = df6['DESCRIPCION'].iloc[m]                    
                break
                
        for n in range (len (df6)):
            cod_sino = int(df6['CLAVE'].iloc[n])
            if (cardiovascular == cod_sino):
                df1['cardiovascular'].iloc[i] = df6['DESCRIPCION'].iloc[n]
                break
                
        for o in range (len (df6)):
            cod_sino = int(df6['CLAVE'].iloc[o])
            if (obesidad == cod_sino):
                df1['obesidad'].iloc[i] = df6['DESCRIPCION'].iloc[o]
                break
                
        for p in range (len (df6)):
            cod_sino = int(df6['CLAVE'].iloc[p])
            if (renal == cod_sino):
                df1['renal_cronica'].iloc[i] = df6['DESCRIPCION'].iloc[p]   
                break
                
        for q in range (len (df6)):
            cod_sino = int(df6['CLAVE'].iloc[q])
            if (tabaquismo == cod_sino):
                df1['tabaquismo'].iloc[i] = df6['DESCRIPCION'].iloc[q]                   
                break
                
        for r in range (len (df6)):
            cod_sino = int(df6['CLAVE'].iloc[r])
            if (otro == cod_sino):
                df1['otro_caso'].iloc[i] = df6['DESCRIPCION'].iloc[r]   
                break
                
        for s in range (len (df6)):
            cod_sino = int(df6['CLAVE'].iloc[s])
            if ( cui== cod_sino):
                df1['uci'].iloc[i] = df6['DESCRIPCION'].iloc[s] 
                break
                    
        for u in range (len(df7)):
            # resultado
            cod_resultado = int(df7['CLAVE'].iloc[u])
            if ( resultado == cod_resultado):
                df1['resultado'].iloc[i] = df7['DESCRIPCION'].iloc[u]
                break
                
        for v in range (len(df8)):
            # nacionalidad
            cod_nacionalidad = int(df8['CLAVE'].iloc[v])
            if (sector == cod_sector):
                df1['nacionalidad'].iloc[i] = df8['DESCRIPCION'].iloc[v]
                break  
                        
lons_lat_dataframe(general, estados, df_sector, df_sexo, df_paciente, df_sino, df_resultado, df_nacionalidad)
```

```python
# importamos librería necesaria para quitar los acentos
import unicodedata
# función que elimina acentos
# Recibe una cadena con tílde
# Devuelve la misma cadena sin tilde
def elimina_tildes(cadena):
    s = ''.join((c for c in unicodedata.normalize('NFD',cadena) if unicodedata.category(c) != 'Mn'))
    return s

# string_acentos = 'café'
# sin_tildes = elimina_tildes(string_acentos)
# print (sin_tildes)
# Tomado de: https://gist.github.com/victorono/7633010
```

```python
# dataframe -> conjunto de datos
# df -> dataframe de datos
def generar_geojson (df):
    
    # crear una lista de features
    lst_features = list()
    for i in range (len(df)):
        
        # iloc para acceder al elemento i de un dataframe
        # Puntos (longitud, latitud) 
        long = int(df['longitud'].iloc[i])
        lat = int(df['latitud'].iloc[i])
        # otras variables
        estado = elimina_tildes(df['estado_um'].iloc[i])
        fecha_ingreso = df['FECHA_INGRESO'].iloc[i]
        fecha_sintomas = df['FECHA_SINTOMAS'].iloc[i]
        fecha_def = df['FECHA_DEF'].iloc[i]
        sector = df['sector'].iloc[i]
        sexo = df['sexo'].iloc[i]
        tipo_paciente = df['tipo_paciente'].iloc[i]
        intubado = df['intubado'].iloc[i]
        neumonia = df['neumonia'].iloc[i]
        edad = int(df['EDAD'].iloc[i])
        diabetes = df['diabetes'].iloc[i]
        epoc = df['epoc'].iloc[i]
        asma = df['asma'].iloc[i]
        inmusupur = df['inmusupr'].iloc[i]
        hipertension = df['hipertension'].iloc[i]
        cardiovascular = df['cardiovascular'].iloc[i]
        obesidad = df['obesidad'].iloc[i]
        renal_cronica = df['renal_cronica'].iloc[i]
        tabaquismo = df['tabaquismo'].iloc[i]
        otro_caso = df['otro_caso'].iloc[i]
        otra_con = df['otra_con'].iloc[i]
        resultado = df['resultado'].iloc[i]
          
        # Geometria de tipo punto
        longlat = Point((long,lat))
        # crear feature
        feature_point = Feature(geometry = longlat, id = i, properties = {
        'state_name': estado, 'fecha_ingreso': fecha_ingreso, 'fecha_sintomas': fecha_sintomas,
        'fecha_defuncion': fecha_def, 'sector': sector, 'sexo': sexo,'tipo_paciente': tipo_paciente,
        'intubado': intubado, 'neumonia': neumonia, 'edad': edad,'diabetes':diabetes,
        'epoc': epoc, 'asma': asma, 'inmusupur': inmusupur, 'hipertension': hipertension,
        'cardiovascular': cardiovascular, 'obesidad': obesidad, 'renal_cronica': renal_cronica,
        'tabaquismo': tabaquismo, 'otro_caso': otro_caso, 'otra_condicion': otra_con,
        'resultado': resultado})
        
        lst_features.append(feature_point)
    # retornando la colección de features
    return FeatureCollection(lst_features)

# llamado de la funcion y le paso el dataframe que tiene los datos
geojson_luis = generar_geojson (general_mod)
```
